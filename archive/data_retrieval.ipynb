```python
import pandas as pd
from datetime import date
from datetime import datetime
from googlesearch import search
from newsplease import NewsPlease
from helper import GoogleSearchQueryOptimizer
```

```python
query = "Joe Biden told people not to vote"
```

```python
today = date.today().strftime('%b %d, %Y')
```

```python
optimizer = GoogleSearchQueryOptimizer()
```

```python
google_search_queries = optimizer(query, today)
```

```python
google_search_queries
```

```python
context = {
    'source': [], 'url': [], 'title': [], 
    'description': [], 'maintext': [], 'date_publish': []
}
```

```python
for i, src in enumerate(['nytimes', 'bbc', 'cnn', 
                         'theguardian', 'washingtonpost']):
    res = search(
        google_search_queries[i],
        advanced=True, 
        num_results=2)

    for r in res:
        context['source'].append(src)
        context['url'].append(r.url)
        context['title'].append(r.title)
        context['description'].append(r.description)  
```

```python
for i in range(len(context['url'])):
    article = NewsPlease.from_url(context['url'][i])
    try:
        context['maintext'].append(article.maintext)
        context['date_publish'].append(article.date_publish)
    except:
        context['maintext'].append("")
        context['date_publish'].append("")
```

```python
context = pd.DataFrame(context)
```
